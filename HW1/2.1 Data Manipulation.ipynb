### 2.1.5 Saving Memory

```python
#Running operations can cause new memory to be allocated to host results
before = id(Y)
Y = Y + X
id(Y) == before #Result of Y + X is allocated in new address, and then points Y to this new location in memory
```

```python
#We want to perform updates in the same location
Z = torch.zeros_like(Y) # have same shape as Y
print('id(Z): ',id(Z))
Z[:] = X + Y # by using Z[:] = <expression> we can assign the result of an operation to an already allocated array Z
print('id(Z): ',id(Z))
```

```python
before = id(X)
X += Y
id(X) == before
```

## Discussions & Exercises

- torch.randn(), creates a tensor with elements drawn from a standard Gaussian (normal) distribution with mean 0 and standard deviation 1
- Indexing starts with 0
- Negative indexing to access an element based on its position relative to the end of the list
- Ranges: X[start:stop], first element(start) included, last(stop) not included
- Only one index for a kth-order tensor is specified, it is applied along axis 0
- Elementwise operators => two tensors of the same shape
- Linear algebraic operations: dot products, matrix multiplications, ...
- Broadcasting:
  - Expand one or both arrays by copying elements along axes with length 1 => both tensors will have the same shape
  - Perform elementwise operation on the resulting arrays
- Tensor class is the main interface for storing and manipulating data in deep learning libraries
